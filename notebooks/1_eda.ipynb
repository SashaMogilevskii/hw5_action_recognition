В data_json и data содержится аналогичная информация. Будем использовать для создания нужного нам датафрейма в дальнейшем data

```python
# Выберем сэмплы, в которых встречается слово "dancing"

df = data[data.label.str.contains('dancing')]

print(data.shape, df.shape)
```

```python
# Посмотрим сколько уникальных классов у нас осталось
df.label.nunique()
```

```python
# Посмотрим на балансировку классов
fig = px.histogram(df, x="label", nbins=len(df["label"].unique()))
fig = fig.update_xaxes(categoryorder='total descending')
fig.update_traces(marker=dict(color='green')) 
fig.update_layout(title_text="Distribution of Labels")

fig.show()
```

```python
# По графику видно, что самый непопулярный таргет является "mosh pit dancing" - 618 значений
# Для чистоты эксперимента и из-за небольшого объему ресурсов возьмем по 200 значений для каждого таргета
# Изначально мы думали взять по 200, но когда мы увидели скорость работы нашего скрипта по скачиваю и парсинга видео - решили что и 200 норм

new_df = pd.DataFrame()
label_values = df['label'].unique()
for label in label_values:
    label_data = df[df['label'] == label].sample(200)
    new_df = pd.concat([new_df, label_data])

new_df.shape, df.shape
```
